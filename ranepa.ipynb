```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = 'https://my.ranepa.ru/online/pk/list/list.php?FT=1&FL=0&FK=&FC=&FB=&FO=&F1=3dfc1680-5eff-4887-a318-288990d0e055&F2=c2d7c59e-e1f0-11e5-80ee-005056a00b6e&F3=fa158c88-0d46-11ef-b514-00155da4ec08&F4=d759df4a-b9a8-11ec-a2da-149ecf4cb24c'


response = requests.get(url)
response.encoding = 'utf-8'

soup = BeautifulSoup(response.text, 'html.parser')

rows = soup.find_all('tr', attrs={'data-npp': True})

snils_numbers = []
for row in rows:
    cells = row.find_all('td')
    if len(cells) > 1:  
        snils = cells[1].get_text(strip=True)
        snils_numbers.append(snils)

ranepa_sociology = snils_numbers

print(snils_numbers)
with open('ranepa_sociology.txt', 'w') as f:
    f.write(' '.join(ranepa_sociology))
```
